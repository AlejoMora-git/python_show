```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df_autos = pd.read_csv("./Vehicle_Price.csv")
df_autos.info()
```

```python
# Adaptar los valores faltantes de descripci√≥n

df_autos['description'] = df_autos['description'].fillna('none')
df_autos
```

```python
# Quitando filas con valores relevantes faltantes

df_autos.dropna(inplace=True)
```

```python
# Nueva composici√≥n del DataFrame compuesto solo de datos completos

df_autos.info()
```

Distribuci√≥n de precios

```python
plt.figure(figsize= (12.8, 7.2), dpi= 150)
sns.histplot(
    data = df_autos,
    x  = 'price',
    bins = 30,
    kde = True,
    color = '#229954'
)
plt.title('Distribucion de precios de veh√≠culos', fontsize = 20)
plt.xlabel('Precio (USD)', fontsize  = 14)
plt.ylabel('Conteo', fontsize = 14);
```

Precio promedio por marca

```python
plt.figure(figsize= (12.8, 7.2), dpi= 150)
precio_promedio = df_autos.groupby('make')['price'].mean().sort_values(ascending = False).head(10)

sns.barplot(
    x = precio_promedio.index,
    y = precio_promedio.values,
    hue = precio_promedio.values,
    palette = 'winter_r'
)

plt.title('Ranking de las 10 Marcas mas caras', fontsize = 20)
plt.ylabel('Precio promedio (USD)', fontsize = 14)
plt.xlabel('Marca', fontsize = 14)
plt.xticks(rotation = 45);
```

An√°lisis de correlacion entre variables

```python
# Definici√≥n de la matriz de correlaci√≥n

corr = df_autos[['price', 'year', 'mileage', 'cylinders', 'doors']].corr()

# Mapa de calor

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de correlaci√≥n');
```

Relaci√≥n entre a√±o  y precio

```python
plt.figure(figsize= (12.8, 7.2), dpi= 150)

sns.boxplot(
    data = df_autos,
    x = 'year',
    y = 'price',
    hue = 'year'
)

plt.title('Relaci√≥n entre a√±o y precio', fontsize = 20)
plt.ylabel('Precio', fontsize = 14)
plt.xlabel('A√±o', fontsize = 14);
```

<li>Los autos mas nuevos (a√±o 2025) no mustran precios mas altos que los a√±os anteriores.

<li>Los autos del a√±o 2024 no solo muestran precios mas altos que en los demas a√±os sino que tienen un rango mayor de precios y muestra los outliers m√°s altos y bajos

Predicci√≥n del precio con Machine Learning

```python
# Preparaci√≥n del modelo

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Selecci√≥n de variables dependientes e independientes para el modelo

caracteristicas = ['year', 'mileage', 'cylinders', 'doors']

X = df_autos[caracteristicas]
y = df_autos['price']

# Divisi√≥n entre entrenamiento y prueba. Fijando random_state para obtener 
# predicciones consistentes a fines de exponer siempre el mismo resultado

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    train_size = 0.8, 
    random_state = 42 )
```

```python
#  Entrenamiento del modelo

from sklearn.metrics import root_mean_squared_error


model = LinearRegression()

model.fit( X_train, y_train )

# Predicci√≥n

y_pred = model.predict(X_test)

# Evaluaci√≥n del modelo de predicci√≥n

print("R¬≤:", r2_score(y_test, y_pred))
print("RMSE:",root_mean_squared_error(y_test, y_pred))
```

<p> El valor R¬≤ de 0.2612 muestra que el 26.12% de la variaci√≥n en los preicios de los autos puede ser explicado por las variables que elegimos ('A√±o', 'millas recorridas', 'cilindros', 'puertas')
<p>
<p> Un RMSE (ra√≠z cuadrada del error cuadr√°tico medio) muestra que nuestro modelo puede equivocarse por USD 14484.29 en exceso o defecto respecto del valor real al momento de hacer una predicci√≥n.

Visualizacion de prediccion vs datos reales

```python
plt.figure(figsize= (12.8, 7.2), dpi= 150)
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title("Predicciones vs valores reales", fontsize = 20)
plt.xlabel("Precio Real", fontsize = 14)
plt.ylabel("Precio Predicho", fontsize = 14);
```
